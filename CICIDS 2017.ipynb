Dataset columns ploted in various types of graphs.

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 8))

# Calculate the sum of 'Total Fwd Packets' and 'Total Backward Packets'
packet_counts = [df[' Total Fwd Packets'].sum(), df[' Total Backward Packets'].sum()]
labels = ['Total Fwd Packets', 'Total Backward Packets']

ax1.pie(packet_counts,
        labels=labels,
        autopct='%1.1f%%',
        colors=['blue', 'red'])
ax1.set_title('Total Fwd Packets vs Total Backward Packets')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate mean values for each label
PortScan_mean_values = df[df[' Label'] == 'PortScan'].mean(numeric_only=True)
benign_mean_values = df[df[' Label'] == 'BENIGN'].mean(numeric_only=True)


plt.figure(figsize=(12, 18))


plt.subplot(2, 1, 1)
sns.barplot(x=PortScan_mean_values.index, y=PortScan_mean_values.values)
plt.xticks(rotation=90)
plt.xlabel("Features")
plt.ylabel("Mean Value")
plt.title("Mean Values of Features for PortScan Traffic")

plt.subplot(2, 1, 2)
sns.barplot(x=benign_mean_values.index, y=benign_mean_values.values)
plt.xticks(rotation=90)
plt.xlabel("Features")
plt.ylabel("Mean Value")
plt.title("Mean Values of Features for BENIGN Traffic")

plt.tight_layout()

plt.show()
```

```python
def plot_feature_distribution(df1, df2, label1, label2, features):
    """
    Plots the distribution of specified features for two dataframes and two labels.

    Args:
        df1 (pd.DataFrame): The first dataframe.
        df2 (pd.DataFrame): The second dataframe.
        label1 (str): The label to filter in the first dataframe.
        label2 (str): The label to filter in the second dataframe.
        features (list): A list of feature names to plot.
    """
    for feature in features:
        plt.figure(figsize=(10, 6))
        sns.histplot(df1[df1[' Label'] == label1][feature], kde=True, color='blue', label=f'{label1} in df1')
        sns.histplot(df2[df2[' Label'] == label2][feature], kde=True, color='red', label=f'{label2} in df2')
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

# Assuming the Friday morning data is loaded into a dataframe named df_morning
# If not, you need to load it first, e.g., df_morning = pd.read_excel("/content/drive/MyDrive/Friday-WorkingHours-Morning.pcap_ISCX.xlsx")

features = [' Destination Port',' Flow Duration',' Total Fwd Packets',' Total Backward Packets','Total Length of Fwd Packets',' Total Length of Bwd Packets']

# Replace df_morning with the actual variable name for your Friday morning dataframe
# For demonstration, I'm using df for both, you should replace df_morning with the correct variable
plot_feature_distribution(df, df, 'BENIGN', 'PortScan', features)
```
